```python
# ЗАДАНИЕ №1
from datetime import datetime

format_1 = "The Moscow Times -Wednesday, October 02, 2002"
format_2 = "he Guardian - Friday, 11 10 13"
format_3 = "Daily News - Thursday, 18 August 1977"

date_1 = datetime.strptime(format_1.replace(",", "").replace(" ", "").split("-")[1], '%A%B%d%Y')
date_2 = datetime.strptime(format_2.replace(",", "").replace(" ", "").split("-")[1], '%A%m%d%y')
date_3 = datetime.strptime(format_3.replace(",", "").replace(" ", "").split("-")[1], '%A%d%B%Y')

print(date_1)
print(date_2)
print(date_3)
```

```python
# ЗАДАНИЕ №2

from datetime import datetime
from datetime import timedelta

def correct_date(date): 
    '''
    функция проверяет корректная ли дата и возвращает True если корректна и False если не корректна
    '''
    try:
        datetime.strptime(date,"%Y-%m-%d")
        return True
    except Exception:
        return False
     
stream = ['2018-22-02', '2018-02-12', '2018-10-02']
        
for st in stream:
    print(correct_date(st))
    
```

```python
# ЗАДАНИЕ №3

from datetime import datetime
from datetime import timedelta

def date_range_(start_date, end_date):
    '''
    Функция, которая возвращает список дат за период от start_date до end_date
    
    '''
    try:  
        start_date_ = datetime.strptime(start_date, "%Y-%m-%d")
        end_date_ = datetime.strptime(end_date, "%Y-%m-%d")
        
        # если start_date больше end_date то функция возвращает пустой список
        if start_date_ > end_date_:
            return []
        else:
            # пустой список для хранения дат
            spisok_data = []
            # первая дата 
            date_sp = start_date_ 
            
            # пока date_sp меньше end_date увеличиваем date_sp на 1 и добавляем к списку date_sp 
            while date_sp < end_date_:
                spisok_data.append(date_sp.strftime("%Y-%m-%d"))
                date_sp += timedelta(days=1)
            return spisok_data
    
    # если дата введена не в формате YYYY-MM-DD, то функция возвращает пустой список 
    except Exception:
        return []

    
print(date_range_("2020-12-22", "2021-12-29"))    
    
```

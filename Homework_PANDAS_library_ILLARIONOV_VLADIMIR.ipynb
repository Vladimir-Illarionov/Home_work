```python
# ЗАДАНИЕ №1

import pandas as pd

data_ratings = pd.read_csv('ratings.csv')
print(data_ratings)

data_movies = pd.read_csv('movies.csv')
print(data_movies)

# найдем номер movieId которому было выставлено больше всего оценок 5.0
number_movieId = data_ratings.query('rating == 5')["movieId"].value_counts().index[0]

# найдем название фильма 
name_film = data_movies.loc[number_movieId]["title"]

print("Больше всего оценок было выставлено фильму", name_film)
```

```python
# ЗАДАНИЕ №2
data_power = pd.read_csv("power.csv")

filtered_countries = data_power[((data_power['country']=="Latvia") | (data_power['country']=="Lithuania") | (data_power['country']=='Estonia'))
                                & ((data_power['category']==4) | (data_power['category']==12) | (data_power['category']==21))
                                &(2005 <= data_power['year']) & (data_power['year'] <= 2010) & (0 <= data_power['quantity'])]


print(filtered_countries)

sum_consumption = filtered_countries["quantity"].sum()

print(sum_consumption)




```

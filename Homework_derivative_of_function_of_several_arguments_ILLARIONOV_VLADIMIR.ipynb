```python
import numpy as np
import sympy as sym
from scipy.optimize import approx_fprime


# Посчитаем градиент функции
def F1(x):
    return 2*(x[0]**2)*(x[1]**3) + 1/x[0] + (x[1]**2)*x[0] + 7

print("Градиент функции в точке (1,2) равен" ,approx_fprime([1,2],F1,[0.00001,0.00001]))


# Посчитаем частные производные функций
x , y = sym.symbols('x y')
F1 = 2*(x**2)*(y**3) + 1/x + (y**2)*x + 7

# Частная производная по x
fx = F1.diff(x)
print("Частная производная по x, F1x=",fx)
# Частная производная по y
fy = F1.diff(y)
print("Частная производная по y, F1y=",fy)


F2= (x**2)*y - sym.sin(x*y) + sym.cos(x**2) + 6*y

# Частная производная по x
fx = F2.diff(x)
print("Частная производная по x, F2x=",fx)
# Частная производная по y
fy = F2.diff(y)
print("Частная производная по y, F2y=",fy)


def f(x):
    return (x[0]**2) + (x[1]**2)

coordinate = np.array([300, 200])
learning_rate = 0.05

for i in range(100):
    # Calculate gradient
    grad = approx_fprime(coordinate,f,[0.00001,0.00001]) 

    coordinate[0] = coordinate[0] - learning_rate*grad[0]
    coordinate[1] = coordinate[1] - learning_rate*grad[1]
    print("x,y = ",coordinate,"grad(x,y) - ",grad, "F(x,y) = ",f(coordinate))
    
print(coordinate)
# print("Значение функции в минимуме: ", f(x))


print("Минимум достигается в точке: ", coordinate)
print("Значение функции в минимуме: ", f(coordinate))
```

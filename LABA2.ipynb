```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

iris = pd.read_csv('iris-write-from-docker.csv')

print(iris)
# поделим данные на тренировочную и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,0:4], iris['class'], 
                                                    train_size=0.5, 
                                                    random_state=42)
                                                                                                                    

```

```python
def plotting_graf(X_s, Y_l):
    
    X_signs = X_s.to_numpy()                                                
    Y_log = Y_l.to_numpy()                                                     
        
    figure, axis = plt.subplots(X_signs.shape[1], X_signs.shape[1], figsize=(30, 30))
    
    for j in range(X_signs.shape[1]):
        for k in range(X_signs.shape[1]):
            for i in np.unique(Y_log):
                
                X = X_signs[np.where(Y_log == i)]
                axis[j,k].scatter(X[:,j], X[:,k], label=i,
                edgecolors='none')
                
                axis[j,k].set_title('Основы анатомии matplotlib')
                axis[j,k].set_xlabel(X_s.columns[j])
                axis[j,k].set_ylabel(X_s.columns[k])
                axis[j,k].set_label(X_s.columns[k])
        
    return
    
#     for ux in axis:
#         for ax in ux:
#             ax.plot([1,2,3,4],[5,6,7,8])
#     return
    
    
plotting_graf(X_test, y_test)
```
